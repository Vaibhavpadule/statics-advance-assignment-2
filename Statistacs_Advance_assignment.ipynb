**Question 1: What is hypothesis testing in statistics?**

**Ans:-**
**Hypothesis testing** in statistics is a method used to make decisions or inferences about population parameters based on sample data. It helps determine whether there is enough evidence in a sample to support or reject a specific assumption about a population.

### Key Components of Hypothesis Testing:

1. **Null Hypothesis ($H_0$)**:

   * This is the default assumption or claim, typically stating that there is no effect or no difference.
   * Example: $H_0$: The mean test score is 70.

2. **Alternative Hypothesis ($H_1$ or $H_a$)**:

   * This is what you aim to support, indicating the presence of an effect or a difference.
   * Example: $H_1$: The mean test score is not 70.

3. **Significance Level ($\alpha$)**:

   * The probability threshold for rejecting the null hypothesis, often set at 0.05 (5%).

4. **Test Statistic**:

   * A value calculated from sample data used to decide whether to reject $H_0$.

5. **P-value**:

   * The probability of obtaining results as extreme as the observed, assuming $H_0$ is true.
   * If the p-value ‚â§ $\alpha$, reject $H_0$.

6. **Conclusion**:

   * Based on the p-value or test statistic, decide whether to reject or fail to reject the null hypothesis.

### Example:

Suppose a factory claims its light bulbs last an average of 1000 hours. A quality control analyst wants to test this claim.

* $H_0$: The average lifespan = 1000 hours
* $H_1$: The average lifespan ‚â† 1000 hours

Using sample data, the analyst performs a test. If the result shows a p-value less than 0.05, they reject $H_0$, concluding the average lifespan is statistically different from 1000 hours.



**Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?**

**Ans:-** The **null hypothesis** and the **alternative hypothesis** are key concepts in statistical hypothesis testing.

---

### **Null Hypothesis (H‚ÇÄ):**

The **null hypothesis** is a statement that there is **no effect**, **no difference**, or **no relationship** between variables. It represents the default or status quo assumption that any observed changes are due to chance or random variation.

* **Example:**
  In a drug effectiveness study:
  **H‚ÇÄ:** The new drug has no effect on blood pressure (i.e., the mean change = 0).

---

### **Alternative Hypothesis (H‚ÇÅ or Ha):**

The **alternative hypothesis** is the statement that **contradicts** the null hypothesis. It represents what the researcher aims to support‚Äîthat there **is** an effect, a difference, or a relationship.

* **Example:**
  Continuing the same study:
  **H‚ÇÅ:** The new drug **does** affect blood pressure (i.e., the mean change ‚â† 0).

---

### **Key Differences:**

| Feature          | Null Hypothesis (H‚ÇÄ)         | Alternative Hypothesis (H‚ÇÅ or Ha)       |
| ---------------- | ---------------------------- | --------------------------------------- |
| Meaning          | No effect or relationship    | There is an effect or relationship      |
| Role             | Default assumption           | What you are trying to prove            |
| Symbol           | H‚ÇÄ                           | H‚ÇÅ or Ha                                |
| Tested Directly? | Yes, using statistical tests | Indirectly (accepted if H‚ÇÄ is rejected) |
| Outcome if p < Œ± | **Reject H‚ÇÄ**                | Support for H‚ÇÅ                          |
| Outcome if p ‚â• Œ± | **Fail to reject H‚ÇÄ**        | Insufficient evidence for H‚ÇÅ            |

---

In summary:

* **H‚ÇÄ** assumes no change.
* **H‚ÇÅ** is what you suspect or want to demonstrate.
* You use statistical tests to decide whether there is enough evidence to **reject H‚ÇÄ** in favor of **H‚ÇÅ**.


**Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.**

**Ans:-** The **significance level** in hypothesis testing, typically denoted by the Greek letter **Œ± (alpha)**, is a threshold used to determine whether the **null hypothesis (H‚ÇÄ)** should be rejected.

---

### üîç **Definition:**

The **significance level (Œ±)** is the **probability of rejecting the null hypothesis when it is actually true**. This is also known as the **Type I error rate**.

Common significance levels are:

* **0.05** (5%)
* **0.01** (1%)
* **0.10** (10%)

---

### üß™ **Role in Hypothesis Testing:**

1. **Set Before the Test:**

   * The researcher chooses a significance level **before** conducting the test, based on how much risk of a false positive (Type I error) they are willing to accept.

2. **Compare with the p-value:**

   * After performing the test, a **p-value** is calculated. This is the probability of observing the test result (or more extreme) if the null hypothesis is true.
   * The decision rule is:

     * If **p-value ‚â§ Œ±**, **reject H‚ÇÄ**
     * If **p-value > Œ±**, **fail to reject H‚ÇÄ**

3. **Interpreting the Result:**

   * If the p-value is small (below Œ±), it suggests that the observed data is unlikely under H‚ÇÄ, so we reject H‚ÇÄ in favor of the **alternative hypothesis (H‚ÇÅ)**.
   * If the p-value is large, the data is not sufficiently unusual, so we do **not reject** H‚ÇÄ.

---

### üìå **Example:**

Suppose a drug company claims that their new medication is more effective than the current one. The null hypothesis is that there is **no difference**.

* They set Œ± = 0.05.
* After testing, they get a p-value of 0.03.

Since **0.03 < 0.05**, they **reject the null hypothesis**, suggesting the new drug may be more effective.

---

### ‚úÖ **Conclusion:**

The significance level is crucial in hypothesis testing because it defines the **boundary for making decisions** under uncertainty. It helps researchers balance the risk of making incorrect conclusions based on sample data.


**Question 4: What are Type I and Type II errors? Give examples of each.**

**Ans:-****Type I and Type II errors** are concepts in **hypothesis testing** in statistics, where we make decisions about a population based on sample data.

---

### üîπ Type I Error (False Positive)

* **Definition**: Rejecting the null hypothesis (**H‚ÇÄ**) when it is actually true.
* **In simple terms**: You think there is an effect or difference, but there isn‚Äôt.

**Example**:
Imagine a court trial:

* **H‚ÇÄ**: The defendant is innocent.
* **Type I error**: The jury finds the defendant guilty when they are actually innocent.

**In medicine**:

* A test says a person has a disease (positive result), but they are actually healthy.

---

### üîπ Type II Error (False Negative)

* **Definition**: Failing to reject the null hypothesis when it is actually false.
* **In simple terms**: You think there is no effect or difference, but there actually is.

**Example**:
Using the same court trial:

* **H‚ÇÄ**: The defendant is innocent.
* **Type II error**: The jury finds the defendant not guilty when they are actually guilty.

**In medicine**:

* A test says a person is healthy (negative result), but they actually have the disease.

---

### üîÅ Summary Table:

| Error Type    | What Happens?                      | Reality          | Decision Made            |
| ------------- | ---------------------------------- | ---------------- | ------------------------ |
| Type I Error  | Reject H‚ÇÄ when H‚ÇÄ is true          | No effect exists | Claim there is an effect |
| Type II Error | Fail to reject H‚ÇÄ when H‚ÇÄ is false | Effect exists    | Claim there is no effect |



**Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.**

**Ans:-** The **Z-test** and **T-test** are both statistical methods used to determine whether there is a significant difference between sample and population means (or between two sample means), but they are used under different conditions.

---

### üîπ **Z-test**

**Used when:**

* The population **standard deviation (œÉ)** is known.
* The **sample size is large** (typically **n ‚â• 30**).
* The data is approximately **normally distributed**.

**Examples of when to use a Z-test:**

* Comparing a sample mean to a known population mean when œÉ is known.
* Comparing proportions between two groups.

---

### üîπ **T-test**

**Used when:**

* The population **standard deviation is unknown**.
* The **sample size is small** (typically **n < 30**).
* The data is approximately **normally distributed**.

There are several types of T-tests:

* **One-sample T-test**: Comparing a sample mean to a known value.
* **Independent two-sample T-test**: Comparing means from two independent groups.
* **Paired T-test**: Comparing means from the same group at different times.

---

### ‚úÖ Summary Table:

| Feature            | **Z-test**                     | **T-test**               |
| ------------------ | ------------------------------ | ------------------------ |
| Population œÉ known | Yes                            | No                       |
| Sample size        | Large (n ‚â• 30)                 | Small (n < 30)           |
| Distribution       | Normal or approx. normal       | Normal or approx. normal |
| Application        | Mean or proportion comparisons | Mean comparisons         |

---

### üîç In Practice:

* **If œÉ is unknown**, and especially if you have a **small sample**, use a **T-test**.
* **If œÉ is known** and/or you have a **large sample**, use a **Z-test**.


**Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.**

```python
#Ans:-
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10         # number of trials
p = 0.5        # probability of success
size = 10000   # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

```python
#Ans:-
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50         # Hypothesized population mean
sigma = 0.5       # Known population standard deviation
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

**Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python
#Ans:-
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 100, 15  # mean and standard deviation
n = 1000  # sample size

data = np.random.normal(mu, sigma, n)

# 2. Calculate the sample mean and standard error
sample_mean = np.mean(data)
standard_error = stats.sem(data)  # standard error of the mean

# 3. Compute the 95% confidence interval
confidence = 0.95
margin = standard_error * stats.t.ppf((1 + confidence) / 2.0, df=n-1)
ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# 4. Plot the histogram with confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.**

```python
#Ans:-
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and visualize the standardized data using a histogram.

    Parameters:
        data (list or numpy array): The dataset to standardize.

    Returns:
        z_scores (numpy array): The standardized Z-scores.
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot histogram
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black')
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores
data = [10, 12, 13, 15, 16, 18, 19, 21, 22, 24]
z_scores = calculate_and_plot_z_scores(data)
print(z_scores)
```
